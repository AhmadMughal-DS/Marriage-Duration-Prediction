```python
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder 
from sklearn.preprocessing import OrdinalEncoder
from sklearn.model_selection import train_test_split
import pickle
import joblib
from sklearn.preprocessing import MinMaxScaler
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler

```

```python
x_train,x_test,y_train,y_test = train_test_split(df.iloc[:,0:5],df.iloc[:,-1],test_size=0.3,random_state=42)
    
```

```python
#fill missing value in num_kids column
trf1 = ColumnTransformer([
    ('num_kids',SimpleImputer(strategy='mean'),[4])
],remainder='passthrough')
```

```python
# column transform of categorical of education man or woman with ordinal order
trf1 = ColumnTransformer([
    ('education_man',OrdinalEncoder(categories=[['Other','Primary','Secondary','Preparatory','Professional']]),[0]),
    ('education_woman',OrdinalEncoder(categories=[['Other','Primary','Secondary','Preparatory','Professional']]),[2])
],remainder='passthrough')
```

```python
# scaling use standardization
trf2 = ColumnTransformer([
    ('income_man_woman',StandardScaler(),[1,3])],
remainder='passthrough')
print(trf2)
```

```python
#model
trf3 = LinearRegression()
```

```python
#pipeline
pipe = Pipeline([
    ('trf1',trf1),
    ('trf2',trf2),
    ('trf3',trf3)
])
```

```python
# Display Pipeline
from sklearn import set_config
set_config(display='diagram')
```

```python
pipe.fit(x_train,y_train)
```

```python
y_predict = pipe.predict(x_test)
print(y_predict)
```

```python
from sklearn.model_selection import cross_val_score
print(cross_val_score(pipe,x_train,y_train,cv=10).mean())
print(cross_val_score(pipe,x_train,y_train,cv=10).max())
print(cross_val_score(pipe,x_train,y_train,cv=10).min())
```
